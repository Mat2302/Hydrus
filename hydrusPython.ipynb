```python
# Run Hydrus-1D without GUI
hydrus_exe = r'C:\Program Files (x86)\PC-Progress\Hydrus-1D 4.xx\H1D_CALC.EXE'
path = r'C:\Program Files (x86)\PC-Progress\Hydrus-1D 4.xx'

try:
    # Run HYDRUS-1D using Popen
    run_hydrus = subprocess.Popen([hydrus_exe], cwd=path, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    stdout, stderr = run_hydrus.communicate()

    # Decode bytes to string using utf-8 encoding
    stdout_str = stdout.decode('utf-8')
    stderr_str = stderr.decode('utf-8')

    # Check if there's any output or error
    if stdout_str:
        print('HYDRUS-1D Standard Output:')
        print(stdout_str)
    if stderr_str:
        print('HYDRUS-1D Standard Error:')
        print(stderr_str)

    # Check the return code
    if run_hydrus.returncode == 0:
        print('HYDRUS-1D executed successfully.')
    else:
        print(f'Error running HYDRUS-1D. Return code: {run_hydrus.returncode}')
except Exception as e:
    print(f'An error occurred: {e}')
```
